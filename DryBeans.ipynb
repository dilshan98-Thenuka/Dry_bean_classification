```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("/content/drive/MyDrive/SCS3201_ML/DryBeanDataset/Dry_Bean_Dataset.xlsx")
df
```

```python
df.shape
```

```python
plt.scatter(df.roundness,df.Class,marker='+',color='red')
```

```python
targetColNames = list(df)
targetColNames.remove('Class')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(df[np.array(targetColNames)],df.Class,train_size=0.95, random_state=42)
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=20000)
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test)
```
